Segmentacion con clusters

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

datos = pd.read_csv('iris.csv')

datos_sepal = datos[['sepal_length', 'sepal_width']]
datos_petal = datos[['petal_length', 'petal_width']]
```

## Seccion 1:
### Visualizar los datos:

```python
pd.plotting.scatter_matrix(datos_sepal)
plt.show()
```

### Crear 2 clusters

```python
def clusterizar(datos, n_clusters):
    kmeans = KMeans(n_clusters, n_init="auto")
    datos_clusters = kmeans.fit_predict(datos)

    # Grfica de pares de variables con clusters
    pd.plotting.scatter_matrix(datos, c=datos_clusters, marker='D', alpha=0.7)
    plt.suptitle(f'Grfica con {n_clusters} clusters' )
    plt.show()
    return datos_clusters

clusterizar(datos_sepal, 2)
```

### Estandarizar y repetir

```python
scaler = StandardScaler()
datos_sepal_scaled = scaler.fit_transform(datos_sepal)
datos_sepal_scaled = pd.DataFrame(datos_sepal_scaled, columns=datos_sepal.columns)

clusterizar(datos_sepal_scaled, 2)
```

### Codo para numero de clusters
Fue utilizado: https://www.themachinelearners.com/k-means/, https://www.geeksforgeeks.org/elbow-method-for-optimal-value-of-k-in-kmeans/
Como referencia para la implementacion del algoritmo de codo

```python
K = range(1, 10)
inertia = []
for k in K:
    kmeans = KMeans(n_clusters = k, random_state = 727, n_init='auto')
    kmeans.fit(datos_sepal_scaled)
    inertia.append(kmeans.inertia_)

plt.plot(K, inertia, 'bx-')
plt.xlabel('Valores de K')
plt.ylabel('Inercia')
plt.title('Mtodo del codo usando inercia')
plt.show()

```

Se nota un codo en el K=3, por lo que se procede a graficar con 3 clusters (Y otros)

### Graficar distintos clusters

```python
n_clusters_list = [2, 3, 4, 5]

for n_clusters in n_clusters_list:
    clusterizar(datos_sepal_scaled, n_clusters)
```
